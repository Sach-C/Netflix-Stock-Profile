This project has an emphasis on visualisations by doing a stock profile on Netflix' stock data (a series of visualisations, studies and analyses that go into different aspects a publicly traded company's data).

For the purposes of this project, I will visualise data for the year, 2017.

```python
# Import libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Read in Netflix data
netflix_stocks = pd.read_csv('NFLX.csv')

# Inspect data
netflix_stocks
```

In the Yahoo financial Data, `Adj Close` represents the adjusted close price adjusted for both dividends and splits. This means that this is the true closing stock price for a given business day.

```python
# Read in Dow Jones data
dowjones_stocks = pd.read_csv('DJI.csv')

# Inspect data
dowjones_stocks
```

```python
# Read in daily Netflix prices by quarter
netflix_stocks_quarterly = pd.read_csv('NFLX_daily_by_quarter.csv')

# Inspect data
netflix_stocks_quarterly.head(10)
```

The Netflix and DJIA stocks contain the OHLC, adjusted close and volume on the first day of every month for 2017.
The Netfilx Stocks Quarterly file shows these values for every business day of the 2017 year, and shows which quarter the data lies in.

The Dow Jones Industrial Average is an industry reflection of the larger stock market.

```python
# Changing name of Adj Close column to Price (as this is the true closing price)
netflix_stocks.rename(columns={'Adj Close': 'Price'}, inplace=True)
```

```python
# Check column name change worked
netflix_stocks.head()
```

```python
# Changing name of Adj Close column to Price on DJI data (as this is the true closing price)
dowjones_stocks.rename(columns={'Adj Close': 'Price'}, inplace=True)

# Check column name change worked
dowjones_stocks.head()
```

```python
# Changing name of Adj Close column to Price on Netflix data (as this is the true closing price)
netflix_stocks_quarterly.rename(columns={'Adj Close': 'Price'}, inplace=True)

# Check column name change worked
netflix_stocks_quarterly.head()
```

```python
# Violin Plot
# Understand the distribution of the Netflix quarterly stock prices for 2017. 
# In which quarter did stock prices fluctuate the most?
plt.figure()
sns.set_style('whitegrid')
sns.set_context('notebook')
ax = sns.violinplot(x='Quarter', y='Price', data=netflix_stocks_quarterly, palette='Reds')
ax.set_title('Distribution of 2017 Netflix Stock Prices by Quarter', fontweight='bold', y=1.05)
plt.ylabel('Closing Stock Price ($)', fontweight='bold')
plt.xlabel('Business Quarters in 2017', fontweight='bold')

plt.savefig('netflix_q.png', dpi=300)
plt.show()
```

There is a steady increase in the median stock price for each Quarter.
Stock prices fluctuated the most in Quarter 3, as can be seen by the large spread.
There is about a USD40 range for each Quarter except for Q3.

The lowest price was just over USD120 which was in Q1, and the highest price was just over USD200 which was in Q4.

```python
# Scatter Plot
# Check the performance of the Earnings per Share (EPS) 
# Plot estimated Yahoo Projected Earnings for the Quarter compared to the Actual earnings for that Quarter


x_positions = [1, 2, 3, 4]
chart_labels = ["1Q-2017","2Q-2017","3Q-2017","4Q-2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]

plt.scatter(x_positions, earnings_actual, color='red', alpha=0.5)
plt.scatter(x_positions, earnings_estimate, color='blue', alpha=0.5)
plt.title('NFLX: Earnings Per Share in Cents', y=1.05, fontweight='bold')
plt.xticks(x_positions, chart_labels)
plt.legend(['Actual', 'Estimate'], bbox_to_anchor=(1.02, 0.55), loc='upper left', borderaxespad=0)
plt.tight_layout()
plt.savefig('netflix_eps.png', dpi=300)
plt.show()
```

In Q2 and Q4, we see purple dots. This means that the actual was equal to the estimated earnings per share (EPS).

In Q1, the actual EPS was slightly higher than the estimate.
In Q2, we saw a large decrease in the actual & estimated EPS.
Q3 saw an increase in both the actual and estimated EPS (but still below Q1's numbers), with the actual slightly less than the estimate.
Q4 saw the actual and estimate EPS rise to slightly higher than Q1's numbers.
The highest EPS was in Q4 for 2017.

```python
# Bar Plots
# Visualise the Earnings and Revenue reported by Netflix

# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q-2017","3Q-2017","4Q-2017", "1Q-2018"]

# Revenue
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars1_x = [t*element + w*n for element
            in range(d)]

# Earnings
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars2_x = [t*element + w*n for element
             in range(d)]

middle_x = [ (a + b) / 2.0 for a, b in zip(bars1_x, bars2_x)]
labels = ["Revenue", "Earnings"]

fig=plt.figure()
plt.bar(bars1_x, revenue_by_quarter, color='#dc143c')
plt.bar(bars2_x, earnings_by_quarter)
plt.title('NFLX: Earnings per Quarter', fontweight='bold', y=1.05)
plt.xticks(middle_x, quarter_labels)
plt.ylabel('Billions ($)')
plt.legend(labels)
plt.savefig('nflx_e_q.png', dpi=300)
plt.show()



```

```python
# Earnings as percentage of Revenue
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]

p = [(i/j)*100 for i, j in zip(earnings_by_quarter, revenue_by_quarter)]
print(p)
```

We have included Q1 for 2018 in the data, to determine the latest trends. Revenue has been increasing for every quarter, as well as earnings.

A very small percentage of the revenue constitutes earnings (2.4% for Q2-2017, 4.3% for Q3-2017, 5.6% for Q4-2017, 7.8% for Q1-2018.

```python
# Line Charts
# Compare Netflix stock to the Dow Jones Industrial Average in 2017

# Left plot Netflix
# ax1 = plt.subplot(total number rows, total number columns, index of subplot to modify)
fig=plt.figure(figsize=(7,5))
ax1 = plt.subplot(1,2,1)
plt.plot(netflix_stocks['Date'], netflix_stocks['Price'], color='red')
ax1.set_title('NFLX', y=1.02, fontweight='bold')
ax1.set_xlabel('Date', fontweight='bold')
ax1.set_ylabel('Stock Price', fontweight='bold')
plt.xticks(rotation=90)

# Right plot Dow Jones
# ax2 = plt.subplot(total number rows, total number columns, index of subplot to modify)
#fig2=plt.figure(figsize=(7,5))
ax2 = plt.subplot(1,2,2)
plt.plot(dowjones_stocks['Date'], dowjones_stocks['Price'])
ax2.set_title('^DJI', y=1.02, fontweight='bold')
plt.xticks(rotation=90)
ax2.set_xlabel('Date', fontweight='bold')
ax2.set_ylabel('Stock Price', fontweight='bold')

plt.subplots_adjust(wspace=0.5)
plt.tight_layout()

plt.savefig('n_vs_d.png', dpi=300)

plt.show()
```

Both Netflix and the Dow Jones increased from the beginning to the end of 2017, with Netflix's price dropping slightly towards the end of the year.
Netflix's share price was far more volatile than the Dow Jones Industrial Average share price over the course of 2017.
The price of is Netflix is significantly lower than the Dow Jones. 

The DJIA represents the health of the broader U.S. economy - so if a company shows similar price movements upwards in 2017, like Netflix did, this is a good sign.
